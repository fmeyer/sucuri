```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loc, scale = 0., 1.
lp_sample = np.random.laplace(loc, scale, 1000)
```

```python
count, bins, ignored = plt.hist(lp_sample, 30, density=True)
x = np.arange(-8., 8., .01)

pdf = np.exp(-abs(x-loc)/scale)/(2.*scale)
plt.plot(x, pdf)
```

```python
gausian = (1/(scale * np.sqrt(2 * np.pi)) *
      np.exp(-(x - loc)**2 / (2 * scale**2)))
plt.plot(x, gausian)
```
